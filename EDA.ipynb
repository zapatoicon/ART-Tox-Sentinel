```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("naija_art_clinic_clean.csv")

df.shape
```

```python
df.info()
```

```python
#Which ART regimen shows highest renal toxicity risk?
renal_risk = (
    df.groupby("art_regimen")["renal_risk_flag"]
    .mean()
    .sort_values(ascending=False)
)

renal_risk.plot(kind="bar", figsize=(10, 6))
plt.title("Renal Toxicity Risk by ART Regimen")
plt.xlabel("ART Regimen")   
plt.ylabel("Mean Renal Risk Flag")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cardiovascular risk by regimen
cv_risk = df.groupby("art_regimen")["cardio_risk_flag"].mean().sort_values(ascending=False)

cv_risk.plot(kind="bar", figsize=(10, 6), color='red')
plt.title("Cardiovascular Toxicity Risk by ART Regimen")
plt.xlabel("ART Regimen")
plt.ylabel("Mean CV Risk Flag")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
